```python
# Plotting functions
from pharmbio.cp import plotting
# Data loading from CPSign output
from pharmbio.cpsign import *

plotting.update_plot_settings(font_scale=1.5)
```

```python
(sign_lvls,error_rates,_,label) = load_calib_stats('output/validation_stats.csv', sep=',')
```

```python
fig = plotting.plot_calibration(sign_vals=sign_lvls,error_rates=error_rates, title=None)
fig.axes[0].get_legend().remove()
```

```python
(sign_vals, median_widths, mean_widths) = load_reg_efficiency_stats('output/validation_stats.csv', sep=',', skip_inf=False)
```

```python
eff_fig = plotting.plot_pred_widths(sign_vals,median_widths,y_label='Median Prediction Interval width')
```

```python
fig.savefig('output/calibration.png',bbox_inches='tight')
eff_fig.savefig('output/efficiency.png',bbox_inches='tight')
```
